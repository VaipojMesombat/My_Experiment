```python
'''
Title Legend:
    plt.legend ("wording",titil_fontsize=20)
    plt.legend (borderpad=3) : default 0.4
    plt.legend (labelspacing=3) default 0.5

'''
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

product1 = [10,13,23,33,22]
product2 = [12,15,20,17,34]
month = [1,2,3,4,5]
Data={"size":20,"color":"yellow","backgroundcolor":"blue"}
Header = {"size":20,"color":"blue"}

# plt.plot (month,product2,color='m',marker='s',linestyle='-.')  
plt.plot (month,product1,"bs-.")                  
plt.plot (month,product2,"ms-.")     # compact string line style
plt.xlabel ("month",Data)
plt.ylabel ("sale amount",Data)         
plt.title("Sale 2020",Header,loc="center")  # "loc" : "right","left","center"

plt.legend (["Printer","Monitor"],title="Product",loc=0,
            fontsize="medium",facecolor="pink",edgecolor='blue',
            borderpad=1,labelspacing=1)
```
