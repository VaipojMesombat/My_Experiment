<p> Change Index column to ['Source', 'Date'] and also chnage data type 'Date' from "object" to "Date" by "parse_dates=['Date']", because when those are index then we will be easy to manage such as sorting, ploting and etc. </p>

<b> Caution !!! </b>
<p> After we change those column to index then you must be use ".loc" for refer to that column. It means that we could not refer to column with df['column name'] </p>

```python
url='https://datahub.io/core/global-temp/r/monthly.csv'
df=pd.read_csv(url, index_col=['Source', 'Date'], parse_dates=['Date'])
df
```

```python
df.info()
```

```python
df.loc['GISTEMP']
```

```python
df.sort_values(['Date', 'Source'], inplace=True)
df
```

<p> Pandas.plot: It will be use index column for axis </p>

```python
df.loc['GISTEMP'].plot()
```

```python
sns.lineplot(data=df.loc['GISTEMP'])
```

```python
dg=df.loc['GISTEMP']
sns.lineplot(x=dg.index, y=dg['Mean'])
```

```python
df.reset_index()
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x='Date', y='Mean', data=df.reset_index(), hue='Source')
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x='Date', y='Mean', data=df.reset_index(), hue='Source', linewidth=.4)
plt.axhline(0, color='green', linestyle='--')
plt.title('Global temperature from 1880-2016')
```
